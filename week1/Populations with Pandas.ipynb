```python
import numpy as np
import pandas as pd
```

```python
hare_df = pop_df["hare"]
hare_df
#用他们的列标签（列名）可以访问列(panada系列)
```

```python
pop_df.hare #使用列标签作为dataframe的属性
```

```python
#常规的数字计算也可用于dataframes或series
print("Mean Hare Population: ", hare_df.mean())
print("Mean Populations: \n")
print(pop_df[["hare","lynx","carrot"]].mean())
print("\n")
print("Standard Deviations: \n")
print(pop_df[["hare","lynx","carrot"]].std())
```

```python
#describe()方法提供了变量的详细描述
pop_df[["hare","lynx","carrot"]].describe()#显示选中三列，并用discribe方法显示详细描述
```

```python
pop_df.describe()#用discribe方法显示该列表的详细描述
```

```python
pop_df[["hare","lynx","carrot"]].corr()
#函数说明：如果由数据框调用corr函数，那么将会计算每个列两两之间的相似度；如果由序列调用corr方法，那么只是该序列与传入的序列之间的相关度
#返回值：DataFrame调用；返回DataFrame；Series调用：返回一个数值型，大小为相关度
```

```python
pop_df.sort_values(by=['hare'])#列表中对于hare的值进行默认升序排序
```

```python
# 寻找所有野兔数量超过5万的实例
hare_above_50K = pop_df.hare>50000
print(hare_above_50K)
print("\n")
print(pop_df[hare_above_50K])
print("\n")
print(pop_df[hare_above_50K].year)
```

```python
# 找出一个动物物种的数量超过5万的所有实例
above_50K = (pop_df["hare"]>50000) | (pop_df["lynx"]>50000)
pop_df[above_50K]
```

```python
pop2 = pop_df.drop("year", axis=1) #如果调用df.mean(axis=1),我们将得到按行计算的均值；如果我们调用 df.drop((name, axis=1),我们实际上删掉了一列，而不是一行
pop2
#axis=0代表往跨行（down)，而axis=1代表跨列（across)
```

```python
poptable = np.array(pop2) #将dataframe(或系列)转换为Numpy数组
poptable
```
